```python
class Parent:
    def __init__(self,a,b,c) -> None:
        self.a=a
        self.b=b
        self.c=c
    def func1(self):
        print("This function is in parent class.")
 
 
 
class Child(Parent):
    def __init__(self, a, b,c) -> None:
        super().__init__(a,b,c)
        self.a=a
        self.b=b
    def func2(self):
        print("This function is in child class.")
 
 
object = Child(1,2,3)
print(object.b)
object.func2()
```

```python
#multiple inheritance

 # Python program to demonstrate
# multiple inheritance
 
# Base class1
class Mother:
    def __init__(self,mothername,fathername) -> None:
        self.mothername = mothername
        self.fathername=fathername
 
    def mother(self):
        print(self.mothername)
    
# Base class2
 
 
class Father:
    def __init__(self,fathername,mothername) -> None:
        self.fathername = fathername
        self.mothername=mothername
 
    def father(self):
        print(self.fathername)
    
    def mother(self):
        print(self.mothername)

# Derived class
 
 
class Son(Mother,Father):
    def __init__(self, mothername,fathername) -> None:
        Father.__init__(self,fathername=fathername,mothername=mothername)
        super().__init__(mothername,fathername)


    son_name="xyz"
    def parents(self):
        self.father()
        print(self.mother())
        print("Father :", self.fathername)
        print("Mother :", self.mothername)
 
 
# Driver's code
s1 = Son("RAM","SITA")
s1.fathername 
s1.mothername
s1.parents()
s1.son_name
```

```python
#multilevel inheritance
# Python program to demonstrate
# multilevel inheritance

# Base class


class Grandfather:

	def __init__(self, grandfathername):
		self.grandfathername = grandfathername

# Intermediate class


class Father(Grandfather):
	def __init__(self, fathername, grandfathername):
		super().__init__(grandfathername)
		self.fathername = fathername

# Derived class


class Son(Father):
	def __init__(self, sonname, fathername, grandfathername):
		super().__init__(fathername, grandfathername)
		self.sonname = sonname

		# invoking constructor of Father class
		

	def print_name(self):
		print('Grandfather name :', self.grandfathername)
		print("Father name :", self.fathername)
		print("Son name :", self.sonname)


# Driver code
s1 = Son('Prince', 'Rampal', 'Lal mani')
print(s1.grandfathername)
s1.print_name()
```
